## 문제5

```python
# 별점 시스템을 만들려고 한다. 따라서 별점에 대한 데이터가 필요하다. 

# 별점은 1,2,3 즉 3개로 만들것이다. 포인트가 95점 이상이면 3점, 85점 이상이면 2점, 나머지는 1점으로 할 것이다. 
# 리뷰데이터를 통해 각 데이터의 별점을 구하시오.
```

```python
# 별점 판단 함수 정의
#테스트
# aplly() df에 적용.
```

```python
#함수 정의

def check_star(point):
    if point >= 95:
        return '***'
    elif point >= 85:
        return '**'
    else :
        return '*'
```

```python
# 함수 test

check_star(70)
```

```python
# df.columns
```

```python
# points column에 aplly

df['points'].apply(check_star)
```

```python
# 새칼럼 생성

df['stars'] = df['points'].apply(check_star)
```

```python
# df.head(5)
```

```python
df['stars'].value_counts()
```

```python
df.loc[df['points']>=95,['points','stars']]
```

```python
# for stars in df['points']:
#     print (stars)
```

```python
for stars in df['points']:
    if stars >= 95:
        df['star_point'] = 3
    elif stars >= 85:
        df['star_point'] = 2
    else:
        df['star_point'] = 1
```

```python
df.head()
```

```python
df.loc[df['points']>=95,['points','star_point']]
```

```python
df['stars'].value_counts()
```
