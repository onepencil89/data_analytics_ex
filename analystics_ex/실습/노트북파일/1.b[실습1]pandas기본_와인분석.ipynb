- 리뷰점수가 95점 이상이고, Australia 또는 New Zealand 에서 만들어진 와인 데이터를 가져오세요

```python
# 리뷰 95점 이상

df['points']>= 95
```

```python
df
```

```python
df['country'] == 'Australia'
```

```python
# Australia 또는 New Zealand 에서 만들어진 와인 데이터
(df['country'] == 'Australia') | (df['country'] == 'New Zealand')
```

```python
# 리뷰 95 이상, 지역 
(df['points']>= 95) & ((df['country'] == 'Australia') | (df['country'] == 'New Zealand'))
```

```python
# 리뷰점수가 95점 이상이고, Australia 또는 New Zealand 에서 만들어진 와인 데이터를 가져오세요
# df.loc[행, 열]
```

```python
df.loc[(df['points']>= 95) & ((df['country'] == 'Australia') | (df['country'] == 'New Zealand')) , :]
```
