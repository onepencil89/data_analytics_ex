---------------------
# 출력 결과: 
array([11, 45, 82])

```python
datas = [[40, 20, 75, 42, 11],
       [30, 39, 92, 21, 45],
       [94, 40, 15, 27, 82]]
arr=np.array(datas)
arr[:,-1]
```

```python
arr = np.array(datas)
arr2 = []
for line in arr:
    arr2.append(line[-1])
np.array(arr2)
```

## 문제 10

------------------------------
array([[ 2,  4,  6,  8, 10],
